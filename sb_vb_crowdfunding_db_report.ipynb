**2. Description of Database** </br>
The database is designed to manage and track crowdfunding campaigns, backers, contributions, rewards, and payment methods. It serves as a structured repository of data that enables the organization to efficiently manage crowdfunding activities and analyze various aspects of campaign success.</br>

The dataset consists of multiple relational tables designed to manage a crowdfunding platform efficiently. The key tables include:</br>
- Campaigns: Stores details about crowdfunding campaigns such as ID, name, organizer, target amount, duration, and status.</br>
- Rewards: Contains different reward types associated with campaigns and their respective contribution thresholds.</br>
- Backers: Holds information about individuals supporting campaigns.</br>
- Contributions: Tracks the financial contributions made by backers to campaigns.</br>
- PaymentMethods: Details the available payment options for contributions.</br>
The dataset is structured using unique IDs for each table entry, ensuring data integrity and easy reference across relations.</br>

***About the Dataset*** </br>
The dataset is designed to simulate a real-world crowdfunding platform, containing structured data on campaigns, backers, contributions, and rewards. It provides a robust foundation for analyzing fundraising trends, user behavior, and the impact of different campaign strategies. The data can be used for generating business insights, optimizing campaign performance, and improving the crowdfunding experience. </br>

This crowdfunding database consists of multiple interrelated tables that store essential details of fundraising campaigns. The key tables include: </br>

- Campaigns: Stores details of all crowdfunding campaigns, including campaign ID, name, organizer, target amount, start and end dates, and current status.

- Backers: Maintains information about contributors who support different campaigns, including their ID, name, email, phone number, and address.

- Contributions: Records donations made by backers for specific campaigns, including contribution ID, campaign ID, backer ID, payment method, organization or individual ID, contribution amount, and date.

- Rewards: Contains data about incentives provided to contributors based on their donation amount, including reward ID, campaign ID, reward description, and minimum contribution amount.

- Reward Claims: Tracks the claims made by backers for rewards, including claim ID, reward ID, backer ID, claim date, and delivery status.

- Payment Methods: Tracks payment methods used for contributions, including payment method ID, method name, and processing fee percentage.

- Organizations: Stores details of organizations contributing to campaigns, including organization ID, name, contact person, email, phone number, address, and tax ID.

- Individuals: Records details of individual contributors, including individual ID, name, email, phone number, address, and anonymity preference.

The database structure ensures streamlined data retrieval, modification, and analysis to support campaign management and business intelligence needs.

-- 15. List the top 5 campaigns with the highest total contributions.</br>

-- 16. Find the number of unique backers contributing to each campaign.</br>
SELECT Campaigns.CampaignName, COUNT(DISTINCT Backers.BackerID) AS UniqueBackers </br>
FROM Contributions </br>
JOIN Campaigns ON Contributions.CampaignID = Campaigns.CampaignID </br>
JOIN Backers ON Contributions.BackerID = Backers.BackerID </br>
GROUP BY Campaigns.CampaignName;</br>
