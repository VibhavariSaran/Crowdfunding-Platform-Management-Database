### **2. Description of Database** </br>
The database is designed to manage and track crowdfunding campaigns, backers, contributions, rewards, and payment methods. It serves as a structured repository of data that enables the organization to efficiently manage crowdfunding activities and analyze various aspects of campaign success.


### **3. About Database**</br>
This crowdfunding database consists of 8 interrelated tables that store essential details of fundraising campaigns. The key tables include:

- Campaigns: Stores details of all crowdfunding campaigns, including campaign ID, name, organizer, target amount, start and end dates, and current status.

- Backers: Maintains information about contributors who support different campaigns, including their ID, name, email, phone number, and address.

- Contributions: Records donations made by backers for specific campaigns, including contribution ID, campaign ID, backer ID, payment method, organization or individual ID, contribution amount, and date.

- Rewards: Contains data about incentives provided to contributors based on their donation amount, including reward ID, campaign ID, reward description, and minimum contribution amount.

- Reward Claims: Tracks the claims made by backers for rewards, including claim ID, reward ID, backer ID, claim date, and delivery status.

- Payment Methods: Tracks payment methods used for contributions, including payment method ID, method name, and processing fee percentage.

- Organizations: Stores details of organizations contributing to campaigns, including organization ID, name, contact person, email, phone number, address, and tax ID.

- Individuals: Records details of individual contributors, including individual ID, name, email, phone number, address, and anonymity preference.



### **4. Database Schema Overview**

The database consists of eight core tables, each serving a specific function. These tables are interconnected using primary keys (PK) and foreign keys (FK) to maintain referential integrity.

**4.1 Campaigns Table**

Purpose: Stores information about crowdfunding campaigns.

Primary Key: CampaignID

Attributes:

* CampaignID (BINARY(6), PK) → Unique identifier for the campaign.

* CampaignName (VARCHAR(255), NOT NULL) → Name of the campaign.

* Organizer (VARCHAR(255), NOT NULL) → Person or entity organizing the campaign.

* TargetAmount (DECIMAL(15,2), NOT NULL) → Fundraising goal.

* StartDate (DATE, NOT NULL) → Start date of the campaign.

* EndDate (DATE, NOT NULL) → End date of the campaign.

* Status (ENUM('Active', 'Completed', 'Cancelled'), NOT NULL) → Status of the campaign.

**4.2 Backers Table**

Purpose: Stores details of individuals supporting campaigns.

Primary Key: BackerID

Attributes:

* BackerID (BINARY(6), PK) → Unique identifier for the backer.

* Name (VARCHAR(255), NOT NULL) → Backer’s full name.

* Email (VARCHAR(255), NOT NULL, UNIQUE) → Unique email for the backer.

* PhoneNumber (CHAR(10), NULL) → Fixed-length phone number.

* Address (TEXT, NULL) → Address of the backer.

**4.3 PaymentMethods Table**

Purpose: Stores available payment options.

Primary Key: PaymentMethodID

Attributes:

* PaymentMethodID (BINARY(6), PK) → Unique identifier for the payment method.

* MethodName (VARCHAR(255), NOT NULL) → Type of payment method (e.g., Credit Card, PayPal).

* ProcessingFeePercentage (DECIMAL(5,2), NOT NULL) → Processing fee percentage.

**4.4 Organizations Table**

Purpose: Stores details of corporate donors contributing to campaigns.

Primary Key: OrganizationID

Attributes:

* OrganizationID (BINARY(6), PK) → Unique identifier for the organization.

* OrganizationName (VARCHAR(255), NOT NULL) → Name of the organization.

* ContactPerson (VARCHAR(255), NULL) → Name of the primary contact.

* Email (VARCHAR(255), NOT NULL, UNIQUE) → Unique email for communication.

* PhoneNumber (CHAR(15), NULL) → Fixed-length phone number.

* Address (TEXT, NULL) → Organization's address.

* TaxID (CHAR(10), NOT NULL, UNIQUE) → Unique tax identification number.

**4.5 Individuals Table**

Purpose: Stores data about individual contributors.

Primary Key: IndividualID

Attributes:

* IndividualID (BINARY(6), PK) → Unique identifier for the individual.

* Name (VARCHAR(255), NOT NULL) → Full name of the contributor.

* Email (VARCHAR(255), NOT NULL, UNIQUE) → Unique email for communication.

* PhoneNumber (CHAR(10), NULL) → Fixed-length phone number.

* Address (TEXT, NULL) → Contributor’s address.

* IsAnonymous (BOOLEAN, NOT NULL, DEFAULT FALSE) → Indicates whether the contribution is anonymous.

**4.6 Contributions Table**

Purpose: Tracks financial contributions made to campaigns.

Primary Key: ContributionID

Foreign Keys:

* CampaignID (BINARY(6), FK) → References Campaigns(CampaignID).

* BackerID (BINARY(6), FK) → References Backers(BackerID).

* PaymentMethodID (BINARY(6), FK) → References PaymentMethods(PaymentMethodID).

* OrganizationID (BINARY(6), FK) → References Organizations(OrganizationID).

* IndividualID (BINARY(6), FK) → References Individuals(IndividualID).

Attributes:

* ContributionID (BINARY(6), PK) → Unique contribution identifier.

* ContributionAmount (DECIMAL(15,2), NOT NULL) → Amount donated.

* ContributionDate (DATE, NOT NULL) → Date of contribution.

**4.7 Rewards Table**

Purpose: Stores information about rewards offered for contributions.

Primary Key: RewardID

Foreign Key:

* CampaignID (BINARY(6), FK) → References Campaigns(CampaignID).

Attributes:

* RewardID (BINARY(6), PK) → Unique reward identifier.

* RewardDescription (TEXT, NOT NULL) → Description of the reward.

* MinimumContribution (DECIMAL(15,2), NOT NULL) → Minimum required contribution.

**4.8 RewardClaims Table**

Purpose: Tracks rewards claimed by backers.

Primary Key: ClaimID

Foreign Keys:

* RewardID (BINARY(6), FK) → References Rewards(RewardID).

* BackerID (BINARY(6), FK) → References Backers(BackerID).

Attributes:

* ClaimID (BINARY(6), PK) → Unique reward claim identifier.

* ClaimDate (DATE, NOT NULL) → Date the reward was claimed.

* DeliveryStatus (ENUM('Pending', 'Shipped', 'Delivered', 'Cancelled'), NOT NULL) → Status of reward delivery.
